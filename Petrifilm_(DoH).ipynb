```python
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.feature import blob_doh
from skimage.color import rgb2gray
from skimage.transform import rotate
```

```python
# Load your image (Change 'your_image.jpg' to your actual file path)
image = io.imread("/content/IMG_0205.jpg")

# Rotate image 90 degrees clockwise
image_rotated = np.rot90(image, k=-1)  # k=-1 rotates 90 degrees clockwise

# Convert image to grayscale
image_gray = rgb2gray(image_rotated)

# Apply DoH blob detection
blobs_doh = blob_doh(image_gray, max_sigma=30, threshold=0.006)

# Plot results
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title("Blob Detection using Determinant of Hessian (DoH) - Rotated 90Â° CW")
ax.imshow(image_rotated, cmap="gray")

# Draw detected blobs
for blob in blobs_doh:
    y, x, r = blob
    c = plt.Circle((x, y), r, color='red', linewidth=2, fill=False)
    ax.add_patch(c)

ax.set_axis_off()
plt.show()
```
